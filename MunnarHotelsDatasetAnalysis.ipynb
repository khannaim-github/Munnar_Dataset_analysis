# Data Cleaning 

```python
df.head()
```

1. Let's first start with the 'Tax' column. We need to perform some pre-processing on Tax column to remove commas and nulls from the column

```python
print(type(df['Tax'].iloc[0]))                 # Data type of record from Tax xolumn
df['Tax'] = df['Tax'].apply(lambda x: str(x))  # Convert each record of Tax column to string format
df['Tax'] = df['Tax'].apply(lambda x: x.replace(',','')) # Replace ',' sign with ''
df['Tax'] = df['Tax'].astype('float64')        # Convert the data type of Tax column to float64
df['Tax'] = df['Tax'].fillna(df['Tax'].median())# Replace the nulls with median values and remove the nulls
df['Tax'].head()                                   # Check the Column 
```

Done with the cleaning of Tax column

2. Let's work on the column 'Star Raating' column

```python
df['Star Rating'].value_counts()       # Check for the value_counts of each entry
df['Star Rating'].fillna(df['Star Rating'].mode().index[0],inplace = True) # remove the duplicates with most repeated value from the column


# Either we can use below code as well in place of above 
#df['Star Rating'].fillna(df['Star Rating'].value_counts().index[0],inplace = True) # remove the duplicates with most repeated value from the column

df['Star Rating'] = df['Star Rating'].astype('int8') # convert the data type of star rating column to int8 form
df['Star Rating'].head()
```

Done with the cleaning of Star Rating column

```python
# Corss check whether null values are removed from the dataset

df.isnull().sum()
```

```python
df.head()
```

3. Clean 'Price' column from the dataset

```python
print(type(df['Price'].iloc[0]))
df['Price']=df['Price'].apply(lambda x: str(x)) # Convert all the elements from Price column into string
df['Price']=df['Price'].apply(lambda x: x.replace(',','')) # Replace ',' with '' from Price column
df['Price']=df['Price'].astype('float64') # Convert the data type of Price column to float64
df['Price'].head()
```

4. Clean 'Distance to Landmark' column

```python
print(type(df['Distance to Landmark'].iloc[0]))
df['Distance to Landmark']=df['Distance to Landmark'].apply(lambda x: str(x))
```

As could not conver the string to float format directly. We need to right one function which will convert the "m" in "km" and even take care of these sign.

```python
def get_dist_in_km(x):
    if x.split()[1] == 'm':
        return float(x.split()[0])/1000                  #converting 'm' into 'km'
    elif x.split()[1] == 'km':
        return float(x.split()[0])
    
# Apply above function on column Distance to Landmark

df['Distance to Landmark']=df['Distance to Landmark'].apply(get_dist_in_km)
```

```python
# convert the Distance to Landmark column into float data type 
df['Distance to Landmark']=df['Distance to Landmark'].astype('float64') 
```

```python
# Create colour pattern function to differentiate the values

def my_map(x,avg):
    if x < avg:
        return 'color: ' + 'Green'
    else:
        return 'color: ' + 'red'
```

```python
# Calculate the Average value of Price column

avg_val=df['Price'].mean()
```

1. Highlight the list of 'Price' with red color having values greater than Avg. Value and with green color having value less than the Avg. value


```python
df[['Price']].style.applymap(lambda x:my_map(x,avg_val))
```

## Univaraite Analysis:

```python
# Import Matplotlib and seaborn libraries to work more on visualization part deepely

import matplotlib.pyplot as plt
import seaborn as sns

for j,i in enumerate(df.columns):
    if df[i].dtype == 'float':
        plt.figure(j)
        sns.displot(df[i],kde=False)
```

### We can justify above points by using below method as well:

```python
# 1. Most of the customer rating to hotels of Munnar between 4 to 4.25.

plt.figure(figsize=(12,8))
plt.title("Hotel Rating at Munnar")
fig_x_y= df.groupby('Rating')['Rating'].count()
sns.barplot(x=fig_x_y.index,y=fig_x_y.values)
plt.show()
```

```python
# 1. You can find hotel in Munnar starting from Rs.2000 and upto Rs.12000. And for most of the hotels rent is between Rs.2000 to Rs.3000

plt.figure(figsize=(12,8))
plt.title("Hotel Price at Munnar")
sns.boxplot(df['Price'],orient='h')
plt.show()
```

```python
# Rating Description:

plt.figure(figsize=(12,8))
plt.title("Rating Description of Hotels")
fig_x_y=df.groupby('Rating Description')['Rating Description'].count()
sns.barplot(x=fig_x_y.index,y=fig_x_y.values)
plt.show()
```

 Most of the hotels got 'Very Good' and 'Excellent' rating.


```python
# Pie Chart

plt.figure(figsize=(12,8))
#plt.title("Rating Description using Pie Chart",loc='left')
fig_x_y=df.groupby('Rating Description')['Rating Description'].count()
plt.pie(fig_x_y,labels=fig_x_y.index,autopct='%.2f',explode=[0,0.1,0,0.1])
plt.show()      
```

1. Out of 100%, 58.1% customers gives 'Very Good' rating to Munnar Hotels.
2. Whereas 33.8% customers gives 'Excellent' rating to Munnar Hotels.
